# Visualization

**Single line Graph**

```python
plt.plot(rs['sepal_length'])
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.title('Line Graph of Sepal Length')
plt.show()
```

**Multiple Line Graph**

```python
plt.plot(rs['sepal_length'], label='Sepal Length')
plt.plot(rs['sepal_width'], label='Sepal Width')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Multiple Line Graph')
plt.legend()
plt.show()
```

**Bar Graph**


```python
rs['species'].value_counts().plot(kind='bar')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar Graph of Species')
plt.show()
```

**Histrogram**

```python
plt.hist(rs['petal_length'])
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

**PIE CHARTS**

```python
rs['species'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Pie Chart of Iris Species')
plt.show()
```

**Scatter Graph**

```python
plt.scatter(rs['sepal_length'], rs['petal_length'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter Plot: Sepal vs Petal Length')
plt.show()
```

```python
x = rs.iloc[:,:-1] # Features
y = rs['species'] # Target variable (species)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
import numpy as np
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```
